```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import re
```

```python
doordash_df = pd.read_csv('Figure1_100Samples/Scores/doordash_job_description.csv')
dd_new_df = pd.read_csv('Figure1_100Samples/Scores/Scores-DD_PM-Resumes.csv')
```

```python
doordash_list_new = ['CV DD_PM Score', 
                     'Claude-3.5-Sonnet DD_PM Score', 
                     'Clean_Deepseek_67B DD_PM Score', 
                     'Clean_Deepseek_V3 DD_PM Score',
                     'GPT-3.5-Turbo DD_PM Score',
                     'GPT4o-mini DD_PM Score', 
                     'GPT4o DD_PM Score',
                     'Llama3-70B-Instruct-Turbo DD_PM Score',
                     'Mixtral-8x7B-Instruct DD_PM Score']

doordash_list =  ['Original CVDoorDash Score',
                  ' claude-3-5-sonnet Conversation-Improved CV DoorDash PM Score',
                  ' Deepseek 67B Conversation-Improved CV DoorDash PM Score',
                  ' GPT-35 Conversation-Improved CV DoorDash PM Score',
                  'GPT-4o-mini Improved CVDoorDash Score',
                  ' GPT-4o Conversation-Improved CV DoorDash PM Score',
                  ' Meta-Llama-3-70B-Instruct-Turbo Conversation-Improved CV DoorDash PM Score',
                  ' Mixtral-8x7B-Instruct-v0.1 Conversation-Improved CV DoorDash PM Score',
                  ]
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.flatten() 
for i, (prev, new) in enumerate(zip(doordash_list, doordash_list_new)): 
    sns.scatterplot(x=doordash_df[prev], 
                   y=dd_new_df[new], 
                   ax=axes[i])
    axes[i].plot([0, 1], [0, 1], transform=axes[i].transAxes, color='red', linestyle='--')
fig.tight_layout()
```
